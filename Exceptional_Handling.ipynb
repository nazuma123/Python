```python
try:
  a = 10
  b = 20
  print(a / b)

except Exception as e:
  print(f'Error Msg given By try Block : {e}')
```

```python
try:
  a = 10
  b = 0
  print(a / b)

except Exception as e:
  print(f'Error Msg given By try Block : {e}')
  error_type , error_msg , error_line = sys.exc_info()
  print(f'Error Type : {error_type}')
  print(f'Error Message by Sys : {error_msg}')
  print(f'Error Line no : {error_line.tb_lineno}')
```
