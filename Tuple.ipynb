**Python Tuples**


- `String -> ''`
- `List -> []`
- `tuple -> ()`


- `If we write anything inside the () -> it comes under tuple data type `
- `we can pass any data type [int | float | string | boolean] inside the ()`
- `Memory Allocation | Slicing | skipping concepts are same like string and list `


```python
a = (10,45.77,'data',False) # save in the Memory

print(a) # calling from Memory
```

```python
a = (10,50,90,100,12,6,3,13)

print(a[4])
```

```python
print(a[2:7])
```

```python
print(a[2 : ])
```

```python
print(a[ : 5])
```

```python
print(a[: : 3])
```

```python
print(a[: : -1])
```

**tuple is mutable or Immutable**

- `try to increase the Memory size `
- `try to decrease the Memory size `


```python
# try to increase

a = (10,50,100) # save in the Memory

a[3] = 500

print(a)
```

```python
# try to decrease

a = (10,50,100) # save in the Memory

del a[1]

print(a)
```

**From the Above operations we can say that Python tuple is Immutable**
