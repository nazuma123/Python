```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# scatter plot
india = [1,5,6,2,7,4]
usa = [5,4,1,3,2,6]
uk = [6,3,5,4,1,7]

plt.figure(figsize = (5,3)) # (w,h)
plt.title('Scatter plot')
plt.xlabel('India')
plt.ylabel('USA')

plt.scatter(x = india , y = usa,color='r',marker='*')

plt.show()
```

```python
# line chart

india = [1,5,6,2,7,4]
usa = [5,4,1,3,2,6]
uk = [6,3,5,4,1,7]

plt.figure(figsize = (5,3)) # (w,h)
plt.title('Scatter plot')
plt.xlabel('India')
plt.ylabel('USA')

plt.plot(india , usa,color='r',marker='*')

plt.show()
```

```python
# line chart

india = [1,5,6,2,7,4]
usa = [5,4,1,3,2,6]
uk = [6,3,5,4,1,7]

plt.figure(figsize = (5,3)) # (w,h)
plt.title('line chart')
plt.xlabel('India')
plt.ylabel('USA')

plt.plot(india , usa,color='r',marker='*')
plt.plot(india , uk,color='b',marker='*')


plt.show()
```

```python
# line chart

india = [1,5,6,2,7,4]
usa = [5,4,1,3,2,6]
uk = [6,3,5,4,1,7]

plt.figure(figsize = (5,3)) # (w,h)
plt.title('line chart')
plt.xlabel('India')
plt.ylabel('USA')

plt.plot(india , usa,marker='*',label='india-usa')
plt.plot(india , uk,marker='*',label='india-uk')

plt.legend()

plt.show()
```

```python
# subplot

plt.figure(figsize=(8,3))

plt.subplot(1,2,1)
plt.title('india-usa')
plt.plot(india,usa,color='r',marker='*')

plt.subplot(1,2,2)
plt.title('india-uk')
plt.plot(india,uk,color='g',marker='*')

plt.show()
```

```python
# subplot

plt.figure(figsize=(5,5))

plt.subplot(2,1,1)
plt.title('india-usa')
plt.plot(india,usa,color='r',marker='*')

plt.figure(figsize=(10,10))
plt.subplot(2,1,2)
plt.title('india-uk')
plt.plot(india,uk,color='g',marker='*')

plt.show()
```

```python
# Histogram

age = [1,4,7,9,3,11,14,21,27,29,31,32,35,39]

plt.figure(figsize=(5,3))

plt.hist(x = age,bins = 4,color='g')

plt.show()
```

```python
# Bar chart

company_names = ['Google','Meta','MSFT','IBM']
revenue = [1200 , 1400,1000,500]

plt.figure(figsize=(5,3))
plt.bar(company_names,revenue,color=['r','g','y','black'])

plt.show()
```

```python
# Bar chart

company_names = ['Google','Meta','MSFT','IBM']
revenue = [1200 , 1400,1000,500]

plt.figure(figsize=(5,3))
plt.barh(company_names,revenue,color=['r','g','y','black'])

plt.show()
```

```python
# pie chart

company_names = ['Google','Meta','MSFT','IBM']
revenue = [1200 , 1400,1000,500]

plt.figure(figsize=(5,3))

plt.pie(revenue,labels=company_names)

plt.show()
```

```python
# pie chart

company_names = ['Google','Meta','MSFT','IBM']
revenue = [1200 , 1400,1000,500]

plt.figure(figsize=(5,3))

plt.pie(revenue,labels=company_names,shadow=True,autopct='%.2f')

plt.show()
```

```python
# pie chart

company_names = ['Google','Meta','MSFT','IBM']
revenue = [1200 , 1400,1000,500]

plt.figure(figsize=(5,3))

plt.pie(revenue,labels=company_names,shadow=True,autopct='%.2f',explode=[0.3,0,0.2,0])

plt.show()
```
