```python
# array = which is used to create data in the form of dimension

f = np.array([10,14,24,35,60]) # 1d

print(f) # calling from Memory

print(f.ndim) # ndim (number of dimensions)

print(f.shape) # shape(rows and columns)
```

```python
print(f[2])
```

```python
t = []

for i in range(11):
  t.append(i)

y = np.array(t)

print(y)
print(y.ndim)
print(y.shape)
```

```python
print(y[6])
```

```python
# 2d

a = np.array([  [10,1,9]  , [3,6,7] , [2,4,8]  ])

print(a)
print(a.ndim)
print(a.shape)
print(a.size) # (total values)
```

```python
print(a[1][2])
```

```python
# changing 1D to 2D
a = np.array([10,11,14,17,19,16])
print(a)
print(a.ndim)
print(a.shape)

b = a.reshape(2,3)
print(b)
print(b.ndim)
print(b.shape)
```

```python
# arange = only 1D data with (n-1) logic

a = np.arange(10,21)
print(a)
print(a.ndim)
```

```python
# 1D to 2D

a = np.arange(1,61)
print(a)
print(a.ndim)
print(a.shape)

b = a.reshape(10,6)
print(b)
print(b.shape)
print(b.ndim)
```

```python
# 1D to 2D

a = np.arange(1,61).reshape(10,6)
print(a)
print(a.ndim)
print(a.shape)
```

```python
print(a[2][1])
```

```python
# 1D to 3D

a = np.arange(1,61).reshape(2,5,6)
print(a)
print(a.ndim)
print(a.shape)
```

```python
print(a[1][2][3])
```

```python
print(a[3][0][4])
```

```python
# 2d to 3D

a = np.array([  [10,1,9]  , [3,6,7] , [2,4,8]  ])
print(a)
print(a.ndim)
print(a.shape)


b = a.reshape(1,3,3)
print(b)
print(b.ndim)
print(b.shape)
```

```python
# numpy -> random -> randint

a = np.random.randint(100,500,3)
print(a)
```

```python
# ravel = will convert any dimension data into 1D

a = np.array([  [10,1,9]  , [3,6,7] , [2,4,8]  ])
print(a)
print(a.ndim)
print(a.shape)

b = np.ravel(a)
print(b)
print(b.ndim)
print(b.shape)
```

```python
# transpose = convert rows to columns and columns to rows

a = np.array([  [10,1,9]  , [3,6,7] , [2,4,8]  ])
print(a)

print('================')

b = np.transpose(a)
print(b)
```

```python
# concatination

a = np.array([ [10,1,9],[ 3  ,6  ,7],[ 2 , 4 , 8] ])
b = np.array([ [10,3 ,2] , [ 1 , 6 , 4] , [ 9  ,7 , 8] ])

print(a)
print(b)
print('================')

c = np.concat((a,b))
print(c)
```

```python
# repeat

a = np.arange(10,21)
b = a.reshape(1,-1)


c = np.repeat(b , 5 , axis=0)
print(c)
```

```python
# some math operations

a = np.array([ [10,1,9],[ 3  ,6  ,7],[ 2 , 4 , 8] ])
b = np.array([ [10,3 ,2] , [ 1 , 6 , 4] , [ 9  ,7 , 8] ])

print(np.add(a,b))
```

```python
# some math operations

a = np.array([ [10,1,9],[ 3  ,6  ,7],[ 2 , 4 , 8] ])
b = np.array([ [10,3 ,2] , [ 1 , 6 , 4] , [ 9  ,7 , 8] ])

print(np.subtract(a,b))
```

```python
# some math operations

a = np.array([ [10,1,9],[ 3  ,6  ,7],[ 2 , 4 , 8] ])
b = np.array([ [10,3 ,2] , [ 1 , 6 , 4] , [ 9  ,7 , 8] ])

print(np.multiply(a,b))
```

```python
# matmul

a = np.array([ [1,4] , [7,6] ])
b = np.array([ [4,3,2] , [1,8,6] ])

print(np.matmul(a,b))
```

```python
# matmul

a = np.array([ [1,4] , [7,6] ])
b = np.array([ [4,3,2] , [1,8,6] , [10,20,30]])

print(np.matmul(a,b))
```
