**varibales**

- `to save the data types in the Memory <temp memory> we are going to use a concept called as varibales `

- `we can store the data type into permanent memory using varibale if we needed But defaulty the varibales will save in the data type in the Temp Memory `

**Rules to create the Varibales**

- `Rule no:1 -> Varibales should not start with a Number if you want to start please use underscore_`

- `Rule no:2 -> Varibale should be either single character or single word multiple charaters with space multiple words with space is not allowed if you want please use underscore `

- `Rule no:3 -> Python Varibales are case sensitive (capital means capital small means small )`

- `Rule no:4 -> Dont use below Punctuations as varibale -> !"#$%&'()*+,-./:;<=>?@[\]^_`{|}~`
