**Comprehensions**

`This concept is used to write the code in single line instead of writing in Multiple Lines `


`There are 2 types of Comprehensions`

      - `List Comprehensions `
      - `Dictionary Comprehensions`
      

**List Comprehensions**


- `Syntax:1 -> one forloop and one print condition`

           [<print_area> <forloop>]


- `Syntax:2 -> One forlopp and one If and one print condition `

            [<Print_Area>     <for loop>   <if_condition>]


- `Syntax:3 -> One forloop one if and one else and there prints `


         [<if_print> <if_condition> <else> <else_print> <forloop>]

```python
for i in range(1,6):
  print('Good Morning')
```

```python
print(['Good Morning' for i in range(1,6)])
```

```python
for i in range(11,16):
  if i%2 == 0:
    print(f'Even : {i}')
```

```python
print([f'Even : {i}' for i in range(11,16) if i%2 == 0])
```

```python
for i in range(11,16):
  if i%2 == 0:
    print(f'Even : {i}')
  else:
    print(f'Odd : {i}')
```

```python
print([f'Even : {i}' if i%2 == 0 else f'Odd : {i}' for i in range(11,16)])
```

**Dictionary Comprehension**

```python
fruits = {'apple':100,'banana':200,'carrot':300}

for i in fruits:
  fruits[i] = fruits[i] * 2

print(fruits)
```

```python
fruits = {'apple':100,'banana':200,'carrot':300}

print({i:fruits[i]*2 for i in fruits})
```

```python
fruits = {'apple':100,'banana':200,'carrot':300}
d = {}
for i in fruits:
  if fruits[i] > 150:
    d[i] = fruits[i] * 3
print(d)
```

```python
fruits = {'apple':100,'banana':200,'carrot':300}

print({i:fruits[i]*3 for i in fruits if fruits[i] > 150})
```

```python
fruits = {'apple':100,'banana':200,'carrot':300}

for i in fruits:
  if fruits[i] > 150:
    fruits[i] = fruits[i] * 3
  else:
    fruits[i] = fruits[i] * 2

print(fruits)
```

```python
fruits = {'apple':100,'banana':200,'carrot':300}

fruits.items()
```

```python
print({i:(j*3 if j>150 else j*2) for i,j in fruits.items()})
```

**calculate Time**

```python
import time
starting_time = time.time()


fruits = {'apple':100,'banana':200,'carrot':300}

for i in fruits:
  if fruits[i] > 150:
    fruits[i] = fruits[i] * 3
  else:
    fruits[i] = fruits[i] * 2

print(fruits)

print(f'Total time : {time.time() - starting_time}')
```

```python
import time
starting_time = time.time()


fruits = {'apple':100,'banana':200,'carrot':300}
print({i:(j*3 if j>150 else j*2) for i,j in fruits.items()})

print(f'Total time : {time.time() - starting_time}')
```
